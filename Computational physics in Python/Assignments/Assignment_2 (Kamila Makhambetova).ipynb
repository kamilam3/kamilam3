```python
import math

def memb1(x,n):
    return (x**n)/math.factorial(n)

def memb2(x,n):
    return ((-x)**n)/math.factorial(n)

def taylser1(x,n):
    y=0
    i=0
    m=-x
    n=n+1
    while i<n:
        y=y+memb2(m,i)
        i=i+1
    return y

def taylser2(x,n):
    m=-x
    y=0
    i=0
    n=n+1
    while i<n:
        y=y+memb1(m,i)
        i=i+1
    return 1/y

trval=math.e**(-5)
i=0

def et(trval,approx):
    return ((trval-approx)/trval)*100

def ea(presapp,prevapp):
    return ((presapp-prevapp)*100)/presapp

print('\n'+'\n'+'\033[1m'+'Table 2: Round-off Errors for 1-x+x^2/2-x^3/3!+....')
print('\n {:12} {:19} {:15} \033[0m'.format('i','e_t%','e_a%'))
k=2
while k<=20:
            print(f'{k:2} {et(trval,taylser1(-5,k)):20.9f} {ea(taylser1(-5,k),taylser1(-5,k-1)):18.9f}')
            k=k+1

            
print('\n'+'\n'+'\033[1m'+'Table 3: Round-off Errors for 1/(1+x+x^2/2+x^3/3!+....)')
print('\n {:10} {:17} {:15} \033[0m'.format('i','e_t%','e_a%'))
k=2
while k<=20:
            print(f'{k:2} {et(trval,taylser2(-5,k)):17.9f} {ea(taylser2(-5,k),taylser2(-5,k-1)):17.9f}')
            k=k+1
```

```python
import numpy as np
import matplotlib.pyplot as plt

def funct(x):
    y = 5*(x**3)-2*x+4
    return y

def fprime(x):
    fprime=15*(x**2)-2
    return fprime
#forward difference of f'(x)
def forwdiff(x,h):
    forw = (funct(x+h)-funct(x))/h
    return forw

#backward difference of f'(x)
def backdiff(x,h):
    back = (funct(x)-funct(x-h))/h
    return back

#central difference of f'(x)
def centdiff(x,h):
    cent = (funct(x+h)-funct(x-h))/(2*h)
    return cent

x=np.linspace(-2,2,17)
h=0.25

y=fprime(x)
fory=forwdiff(x,h)
backy=backdiff(x,h)
centy=centdiff(x,h)

plt.title('Finite difference approximations of f(x) ')

plt.plot(x,y,'black', label=r'$f^\prime(x)$',linewidth=3.0)
plt.plot(x,fory,'blue', label=r'$f^\prime_F(x)$')
plt.plot(x,backy,'green', label=r'$f^\prime_B(x)$')
plt.plot(x,centy,'yellow', label=r'$f^\prime_C(x)$')

plt.xlabel(r'$x$',size=13)
plt.ylabel(r'$y(x)$',size=13)
plt.legend(loc='lower left',fontsize='10',frameon=False)
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    f = 5*(x**3)-2*x+4
    return f

def secder(x):
    secder=30*x
    return secder

#second forward difference of f'(x)
def secforw(x,h):
    forw = (f(x+2*h)-2*f(x+h)+f(x))/(h**2)
    return forw

#second backward difference of f'(x)
def secback(x,h):
    back = (f(x)-2*f(x-h)+f(x-2*h))/(h**2)
    return back

#central difference of f'(x)
def seccent(x,h):
    cent = (f(x+h)-2*f(x)+f(x-h))/(h**2)
    return cent

x=np.linspace(-2,2,17)
h=0.25

y=secder(x)
fory=secforw(x,h)
backy=secback(x,h)
centy=seccent(x,h)

plt.title('Second order finite difference approximations of f(x) ')

plt.plot(x,y,'black', label=r'$f^{\prime\prime}(x)$',linewidth=3.0)
plt.plot(x,fory,'blue', label=r'$f^{\prime\prime}_F(x)$')
plt.plot(x,backy,'green', label=r'$f^{\prime\prime}_B(x)$')
plt.plot(x,centy,'yellow', label=r'$f^{\prime\prime}_C(x)$')

plt.xlabel(r'$x$',size=13)
plt.ylabel(r'$y(x)$',size=13)
plt.legend(loc='lower right',fontsize='10',frameon=False)
plt.grid(True)
plt.show()
```
